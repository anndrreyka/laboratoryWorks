```python
from scipy.interpolate import CubicSpline 

import numpy as np 

import matplotlib.pyplot as plt 

 

# Задані табличні дані 

x = np.array([0.3, 0.8, 1.4, 2.1, 2.9]) 

y = np.array([5.28, 6.83, 4.35, 3.26, 2.35]) 

 

# Побудова кубічного сплайну 

cs = CubicSpline(x, y) 

 

# Генерація значень для плавної кривої 

x_vals = np.linspace(x[0], x[-1], 100) 

y_cs = cs(x_vals) 

 

# Графік табличних даних та кубічного сплайну 

plt.figure(figsize=(8, 6)) 

plt.scatter(x, y, label='Табличні дані', color='red') 

plt.plot(x_vals, y_cs, label='Кубічний сплайн', color='blue') 

plt.title('Кубічний сплайн для табличних даних') 

plt.xlabel('X') 

plt.ylabel('Y') 

plt.legend() 

plt.grid(True) 

plt.show() 
```
