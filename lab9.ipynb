```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 5 * np.cos(x)

def f_prime(x):
    return -5 * np.sin(x)

def f_double_prime(x):
    return -5 * np.cos(x)

def f_triple_prime(x):
    return 5 * np.sin(x)

def P_0(x):
    return 5

def P_1(x):
    return 5

def P_2(x):
    return 5 - (5 / 2) * x**2

def P_3(x):
    return 5 - (5 / 2) * x**2

x_values = np.linspace(-1, 1, 100)

plt.figure(figsize=(10, 6))

plt.plot(x_values, f(x_values), label='f(x)', linewidth=2)
plt.plot(x_values, np.vectorize(P_0)(x_values), label='P_0(x)', linestyle='--')
plt.plot(x_values, np.vectorize(P_1)(x_values), label='P_1(x)', linestyle='--')
plt.plot(x_values, np.vectorize(P_2)(x_values), label='P_2(x)', linestyle='--')
plt.plot(x_values, np.vectorize(P_3)(x_values), label='P_3(x)', linestyle='--')

plt.legend()
plt.title('    ')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
