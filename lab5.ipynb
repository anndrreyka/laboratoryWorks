```python
# Побудова графіка функції 

import numpy as np 

import matplotlib.pyplot as plt 

 

# область значень для x та y 

x_min, x_max = -3, 6 

y_min, y_max = -3, 3 

step = 0.001 

 

# створюємо масиви значень x та y 

x, y = np.meshgrid(np.arange(x_min, x_max, step), 

                   np.arange(y_min, y_max, step)) 

 

# рівняння системи 

eq1 = np.sin(y) + 2 * x - 2 

eq2 = np.cos(x - 1) + y - 0.7 

 

# створюємо графік 

fig, ax = plt.subplots(figsize=(10, 10)) 

 

# додаємо графік першого рівняння 

ax.contour(x, y, eq1, levels=[0], colors='green') 

 

# додаємо графік другого рівняння 

ax.contour(x, y, eq2, levels=[0], colors='purple') 

 

# налаштування графіка 

ax.set_xlim([x_min, x_max]) 

ax.set_ylim([y_min, y_max]) 

ax.set_xlabel('x') 

ax.set_ylabel('y') 

ax.set_title('Графік') 

plt.grid(True) 

 

# показуємо графік 

plt.show() 

 

 

# Розв'язок системи рівнянь методом простої ітерації 

import numpy as np 

import math 

 

x0 = 0 

y0 = 0 

delta = 0.1 

 

def f1(y): 

    return 2 - math.sin(y) / 2   # задаємо функції 

 

def f2(x): 

    return 0.7 - math.cos(x - 1)   # задаємо функції 

 

def iter(x, y, e): 

    xn = x 

    yn = y 

    xn1 = f2(x) 

    yn1 = f1(y) 

    n = 1 

    while ((abs(xn1 - xn) >= e) & (abs(yn1 - yn) >= e)): 

        xn = xn1 

        yn = yn1 

        xn1 = f2(xn)  # змінено з f2(yn) 

        yn1 = f1(yn)  # змінено з f1(xn) 

        n += 1 

    print('Simple iteration:') 

    print('x =', xn, '\ny =', yn, '\nThe amount of iteration = ', n) 

 

iter(x0, y0, 0.0001) 
```
